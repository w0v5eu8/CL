#Data 생성

```python
import numpy as np
import matplotlib.pyplot as plt

X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()
```

#Sci-kit Learn: Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)

print("가중치(계수, 기울기 파라미터 W) :", lin_reg.coef_)
print("편향(절편 파라미터 b) :", lin_reg.intercept_)
```

```python
X_new = np.array([[0], [2]]) # x=0, x=2인 두 곳의 y 값을 찾기 위해서
y_predict=lin_reg.predict(X_new)

print('X: ')
print(X_new)
print('Y: ')
print(y_predict)

plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

#Pseudo Inverse

이 함수는 $\mathbf{X}^+\mathbf{y}$을 계산합니다. $\mathbf{X}^{+}$는 $\mathbf{X}$의 _유사역행렬_ (pseudoinverse)입니다(Moore–Penrose 유사역행렬입니다). `np.linalg.pinv()`을 사용해서 유사역행렬을 직접 계산할 수 있습니다:

$\boldsymbol{\hat{\theta}} = \mathbf{X}^{-1}\hat{y}$

```python
X_b = np.c_[X, np.ones((100, 1))]  # 모든 샘플에 bias term 추가

W = np.linalg.pinv(X_b).dot(y)

print("가중치(계수, 기울기 파라미터 W) :", W[0])
print("편향(절편 파라미터 b) :", W[1])
```

```python
X_new = np.array([[0], [2]]) # x=0, x=2인 두 곳의 y 값을 찾기 위해서
X_new_b = np.c_[X_new, np.ones((2, 1))]  # 모든 샘플에 bias term 추가

y_predict = X_new_b.dot(W)

print('X: ')
print(X_new_b)
print('Y: ')
print(y_predict)

plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

#정규 방정식

$\hat{\boldsymbol{\theta}} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}$

```python
X_b = np.c_[X, np.ones((100, 1))]  # 모든 샘플에 bias term 추가

W = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

print("가중치(계수, 기울기 파라미터 W) :", W[0])
print("편향(절편 파라미터 b) :", W[1])
```

$\hat{y} = \mathbf{X} \boldsymbol{\hat{\theta}}$

```python
X_new = np.array([[0], [2]]) # x=0, x=2인 두 곳의 y 값을 찾기 위해서
X_new_b = np.c_[X_new, np.ones((2, 1))]  # 모든 샘플에 bias term 추가

y_predict = X_new_b.dot(W)

print('X: ')
print(X_new_b)
print('Y: ')
print(y_predict)

plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```
